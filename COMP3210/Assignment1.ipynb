```python
## crete a dataframe of the documents from the Tweets collection in MongoDB
data = pd.DataFrame(list(collection.find()))
data
```

```python
data1 = data['text']
data1.head(10)
```

```python
### importing cleaning libraries
import re # allows use of Regular Expressions (i.e. to keep letters and numbers)
import nltk # most common and easy to use data pre-processing library
from nltk.stem.porter import PorterStemmer # stemming is the process of producing
# morphological variants of a root/base word
nltk.download('stopwords') # standard stopwords list
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize

corpus = []

for i in range(0, len(data1)):
    review = re.sub('[^a-zA-Z0-9]', ' ', data1[i]) ## keeping letters and numbers
    review = review.lower() ## converting to lowercase
    review = review.split() ## split each sentence into individual words for each record
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
new_list = []
for row in corpus:
    new_list.append(row.split(" "))
new_list
```

```python
import csv #libary to export dataframes to csv file

with open("out.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(new_list)
```
